# Aykırı Değer (Outlier) Analizi


# 1. IQR (Interquartile Range) Yöntemi

Bu yöntem, verilerin çeyrek değerlerini kullanarak aykırı değerleri tespit eder. Aykırı değerler, aşağıdaki formüle göre belirlenir:

Alt sınır: Q1 - 1.5 * IQR
Üst sınır: Q3 + 1.5 * IQR

```python
# Aykırı değer analizi ve tablo şeklinde çıktı
def outlier_summary(df):
    summary = []
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
    for column in numeric_columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Aykırı değerlerin tespiti
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outlier_count = len(outliers)
        outlier_percentage = (outlier_count / len(df)) * 100
        
        # Sütun için özet
        summary.append({
            "Column": column,
            "Outlier Count": outlier_count,
            "Outlier Percentage (%)": round(outlier_percentage, 2)
        })
    
    # Tabloyu DataFrame olarak döndürme
    return pd.DataFrame(summary)

# Aykırı değerlerin özet tablosu
outlier_table = outlier_summary(df)
print(outlier_table)
```

# 2. Z-Score Yöntemi

Z-Score, bir değerin ortalamadan kaç standart sapma uzaklıkta olduğunu hesaplar. 
Belirli bir eşik değerden büyük Z-Score değerleri aykırı olarak kabul edilir (örneğin, |z| > 3).

```python
from scipy.stats import zscore

# Z-Score ile aykırı değer analizi
def zscore_outliers(df, threshold=3):
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    outlier_summary = []

    for column in numeric_columns:
        z_scores = zscore(df[column])
        outliers = df[(abs(z_scores) > threshold)]
        outlier_count = len(outliers)
        outlier_percentage = (outlier_count / len(df)) * 100

        outlier_summary.append({
            "Column": column,
            "Outlier Count": outlier_count,
            "Outlier Percentage (%)": round(outlier_percentage, 2)
        })

    return pd.DataFrame(outlier_summary)

# Z-Score yöntemiyle aykırı değer analizi
zscore_table = zscore_outliers(df)
print(zscore_table)
```

# Clustering (DBSCAN) Yöntemi:

DBSCAN algoritması, aykırı değerleri tespit etmek için kullanılabilir. Bu yöntem, veri noktalarını kümelere ayırır ve kümelere dahil olmayan noktaları aykırı değer olarak işaretler.

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import pandas as pd

# Sayısal değişkenleri seçme
numeric_columns = ['Age', 'Num_Interactions', 'Feedback_Score', 'Products_Purchased', 
                   'Products_Viewed', 'Time_Spent_on_Site', 'Satisfaction_Score']

X = df[numeric_columns].values

# Veriyi ölçeklendirme
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# eps değerini belirlemek için Knee/Elbow Method
neigh = NearestNeighbors(n_neighbors=5)
nbrs = neigh.fit(X_scaled)
distances, indices = nbrs.kneighbors(X_scaled)

# Mesafeleri sırala ve grafiği çiz
distances = np.sort(distances[:, 4])
plt.figure(figsize=(8, 5))
plt.plot(distances)
plt.title('Knee/Elbow Method for eps')
plt.xlabel('Data Points sorted by distance')
plt.ylabel('4th Nearest Neighbor Distance')
plt.show()

# Optimal eps değerini belirleme
from kneed import KneeLocator
kneedle = KneeLocator(range(len(distances)), distances, S=1.0, curve="convex", direction="increasing")
optimal_eps = distances[kneedle.knee]
print("Optimal eps:", optimal_eps)

# DBSCAN algoritmasını uygulama
min_samples = 5  # Küçük kümeler için minimum örnek sayısı
dbscan = DBSCAN(eps=optimal_eps, min_samples=min_samples)
dbscan.fit(X_scaled)

# Aykırı değerleri işaretleme
df['DBSCAN_Outlier'] = (dbscan.labels_ == -1).astype(int)
print("Aykırı Değer Sayısı:", df['DBSCAN_Outlier'].sum())

# Aykırı değerlerin görselleştirilmesi (PCA ile 2D indirgeme)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='DBSCAN_Outlier', data=df, palette={0: 'blue', 1: 'red'})
plt.title('DBSCAN ile Aykırı Değer Tespiti (PCA ile 2D)')
plt.show()
```

* min_samples: Küçük kümeler için minimum örnek sayısı olarak ayarlanır (örneğin, 5)
* eps değeri, bir veri noktasının çevresindeki diğer veri noktalarını "komşu" olarak kabul etmek için kullanılan mesafe sınırıdır.
